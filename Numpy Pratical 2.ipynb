```python
# Mathematic Operation Using Numpy
```

```python
import numpy as np
```

```python
arr1 = np.arange(1,10).reshape(3,3)
arr2 = np.arange(1,10).reshape(3,3)
print(arr1)
print(arr2)
```

```python
np.add(arr1,arr2)
```

```python
np.subtract(arr1,arr2)
```

```python
np.divide(arr1,arr2)
```

```python
np.multiply(arr1,arr2)
```

```python
arr1@arr2 #matrix product
```

```python
arr1.max()
```

```python
arr1.argmax() #index
```

```python
arr1.max(axis=0) #column
```

```python
arr1.max(axis=1) #row
```

```python
arr1.min()
```

```python
arr1.argmin()
```

```python
arr1.min(axis=0)
```

```python
arr1.min(axis=1)
```

```python
np.sum(arr1)
```

```python
np.sum(arr1,axis=0)
```

```python
np.sum(arr1,axis=1)
```

```python
np.mean(arr1)
```

```python
np.sqrt(arr1)
```

```python
np.std(arr1)
```

```python
np.exp(arr1)
```

```python
np.log(arr1)
```

```python
np.log10(arr1)
```
