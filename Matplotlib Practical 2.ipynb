```python
import matplotlib.pyplot as plt

# Labels
classes = ['Py', 'R', 'ML', 'AI', 'DS']

# Values
class1_students = [45, 15, 35, 25, 30]

# Explode one slice
explode = [0, 0, 0.2, 0, 0]

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    class1_students,
    labels=classes,
    explode=explode,
    autopct='%0.1f%%',
    shadow=True,
    radius=1.4,
    startangle=90
)

plt.show()
```

```python
#subplot
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
plt.pie([1])

plt.subplot(2, 2, 2)
plt.pie([1, 2])

plt.subplot(2, 2, 3)
plt.pie([1, 2, 3])

plt.subplot(2, 2, 4)
plt.pie([2, 3, 4, 5])

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))

plt.subplot(3, 2, 1, projection='polar', facecolor='k', frameon=True)
plt.subplot(3, 2, 2, projection='polar', facecolor='k', frameon=True)
plt.subplot(3, 2, 3, projection='polar', facecolor='k', frameon=True)
plt.subplot(3, 2, 4, projection='polar', facecolor='k', frameon=True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Read image
img = mpimg.imread("pie_chart.png")

# Image details
print(type(img))
print(img.shape)
print(img.ndim)

# Show image
plt.figure(figsize=(16, 9))
plt.imshow(img)
plt.axis("off")

plt.show()
```
